```python
import math
#definir quantos números serão inseridos na calculadora (array que termina o loop ao apertar o botão de igual)
numbers = []
operations = []
priorities = []
finish = ""

#função importante para saber como prosseguir após do cálculo de uma potência, raíz, multiplicação ou divisão
#substituir o número selecionado pelo número calculado
def gathering(number, beforeOp="", afterOp=""):
    if(beforeOp == "" and afterOp == ""):
        print('calculo único')
    elif (beforeOp != "" and afterOp != ""):
        print(beforeOp, afterOp, number)
    elif (beforeOp != ""):
        print(beforeOp)
    else:
        print(afterOp)
    

def soma(x, y):
    return print(int(x) + int(y))
    
def sub(x, y):
    return print(int(x) - int(y))
    
def div(x, y):
    return print(int(x) / int(y))
    
def mult(x, y):
    return print(int(x) * int(y))
    
def pot(x, y):
    return print(int(x) ** int(y))

def root(x):
    return math.sqrt(int(x))

#função disparada ao pressionar "igual", junta todos os resultados e apresenta o 
# resultado final
def equalsTo():
    cp_priorities = priorities [:] #cópia do array original, pois este será alterado
    priorities.sort(reverse = True)
    #fazer um loop que pega as operações de prioridades, faz a operação e retorna um valor
    for p in priorities:
        if operations[cp_priorities.index(p)] == "//":
            gathering(root(numbers[cp_priorities.index(p)]), operations[cp_priorities.index(p) - 1], operations[cp_priorities.index(p) + 1])
            operations.pop(cp_priorities.index(p))
            priorities.remove(p)
            cp_priorities.remove(p)
            print(operations, priorities, cp_priorities)
            continue
        elif operations[cp_priorities.index(p)] == "**":
            pot(numbers[cp_priorities.index(p)], numbers[cp_priorities.index(p) + 1])
            continue
        elif operations[cp_priorities.index(p)] == "*":
            mult(numbers[cp_priorities.index(p)], numbers[cp_priorities.index(p) + 1])
            continue
        elif operations[cp_priorities.index(p)] == "/":
            div(numbers[cp_priorities.index(p)], numbers[cp_priorities.index(p) + 1])
            continue
        elif operations[cp_priorities.index(p)] == "+":
            soma(numbers[cp_priorities.index(p)], numbers[cp_priorities.index(p) + 1])
            continue
        elif operations[cp_priorities.index(p)] == "-":
            sub(numbers[cp_priorities.index(p)], numbers[cp_priorities.index(p) + 1])
            continue
        print(numbers[cp_priorities.index(p)], operations[cp_priorities.index(p)])
        cp_priorities.pop(cp_priorities.index(p))
    

def setPriority(operator):
    if operator == "**" or operator == "//":
        priorities.append(2)
    if operator == "*" or operator == "/":
        priorities.append(1)
    if operator == "+" or operator == "-":
        priorities.append(0)

while (finish == ""):
    x = input()
    #caso onde o usuário pode digitar somente a raiz quadrada, um número e logo em seguida o operador "="
    if x == "=": equalsTo()
    numbers.append(x)
    op = input()
    operations.append(op)
    setPriority(op)
    
    if op == "=":
        operations.pop()
        equalsTo()
        break
        
    continue
```
